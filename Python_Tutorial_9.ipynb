# Write a program to implement following. 
    Create a base class named person consisting of name and code. Create 2 child classes
    a) Account with member_pay
    b) Admin with experience and inherit the base class.
    Create a class employee with name, code, experience and pay by inheriting the above classes.

```python
class person:
    
    def __init__(self):
        self.name = input("Enter Employee Name :")
        self.code = input("Enter Employee Code :")
        
    def name_code(self):
        print("Employee Name :",self.name)
        print("Employee Code :",self.code)
        
class account(person):
    
    def __init__(self):
        self.payment = int(input("Enter the Payment Amount :"))
    
    def member_pay(self):
        print("Employee Payment is :",self.payment)
        
class admin(person):
    
    def __init__(self):
        self.experiance = input("Enter Employee Experiance in years:")
        
    def member_experience(self):
        print("Employee Experiance :",self.experiance)
        
class employee(admin,account):
    
    def __init__(self):
        person.__init__(self)
        admin.__init__(self)
        account.__init__(self)
        
        
obj = employee()
obj.name_code()
obj.member_experience
obj.member_pay()
```
